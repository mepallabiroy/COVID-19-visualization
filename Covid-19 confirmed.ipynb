# Converting the DataFrame to array

```python
df = dataset.values
df
```

```python
df.shape
```

```python
type(df)
```

# Deleting an column in the array if 70% of the data in the column is 0

```python
lst = []
for i in range(61):
    count = 0
    for j in range(171):
        if df[j,i] == 0:
            count+=1
    if count > 0.7*171:
        lst.append(i)
for i in lst:
    X = np.delete(df, np.s_[:i], 1)
```

```python
X.shape #36 columns are deleted
```

```python
#Converting X to DataFrame

X=pd.DataFrame(X)
X 
```

# Plotting the data

```python
plt.plot(X)
plt.xticks(rotation = 45)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

```python
sns.barplot(data=X)
plt.xticks(rotation = 45)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```
